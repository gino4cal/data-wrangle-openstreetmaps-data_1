# Final XML to JSON Export to MONGOdb

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import xml.etree.ElementTree as ET
from collections import defaultdict
import pprint
import re
import codecs
import json


lower = re.compile(r'^([a-z]|_)*$')
lower_colon = re.compile(r'^([a-z]|_)*:([a-z]|_)*$')
problemchars = re.compile(r'[=\+/&<>;\'"\?%#$@\,\. \t\r\n]')

street_type_re = re.compile(r'\b\S+\.?$', re.IGNORECASE)
expected = ["Street", "Avenue", "Boulevard", "Drive", "Court", "Place", "Square", "Lane", "Road", 
            "Trail", "Parkway", "Commons"]

mapping = { "St": "Street",
            "St.": "Street",
            "Ave": "Avenue",
            "Rd." : "Road",
            "Rd" : "Road",
            "Blvd." : "Boulevard",
            "Blvd" : "Boulevard",
            "Dr" : "Drive"
            }

CREATED = [ "version", "changeset", "timestamp", "user", "uid"]

def update_name(name, mapping):   #this will update the street names based on the mapping dict

    
    m = street_type_re.search(name)  # pick out the last word of the street text
    if m <> None:
        street_type = m.group()  #save text to variable
        if street_type in mapping:
            corrected_street_type = mapping[street_type]   #correct text based on mapping dict
    
            name = name[0:name.find(street_type)] + corrected_street_type   # paste together with corrected text


    return name


def shape_element(element):
    node = {}
    created = {}
    pos = []
    address = {}
    gnis = {}
    '''NOTE:  I used 2 conditional paths between the top elements 'node' and 'way' because there were particular rules
    applied to each.  Though possibly duplicative, this ensured that all conditions were met.'''
    
    if element.tag == "node":
        
        #iterate for all child tags under top element
        for n in element.iter():
            
            #only 'tag' tags evaluated
            if n.tag == 'tag':
                #assign the k and v values of the tag
                key = n.attrib.values()[0]
                
                val = n.attrib.values()[1]
                
                
                #only work on k values with no problem characters, ignore the rest
                if problemchars.search(key) == None:
                    
                    #create the address dictionary, ignore keys with more than 1 colon    
                    if key.startswith('addr:') and key.count(':') < 2:
                        col = key.find(':')
                        address[key[(col + 1):]] = val
                        
                    #create USGS GNIS dictionary, ignore keys iwth more than 1 colon
                    elif key.startswith('gnis:') and key.count(':') < 2:
                        col = key.find(':')
                        gnis[key[(col + 1):]] = val
                        
                    
                    #add the rest of the key pairs to the node
                    elif key.count(':') < 2:
                        node[key] = val
        
        #iterate over the attributes of the main element 'node'
        for k,v in element.attrib.iteritems():
            
            #create the 'created' dictionary
            if k in CREATED:
                created[k] = v
            #create the 'pos' list and make them floats
            elif k == 'lat':
                pos.insert(0,float(v))
            elif k == 'lon':
                pos.insert(1,float(v)) 
            #add the rest of the attribute key,value pairs to the node
            else:
                node[k] = v
        
        #add address dict to the node, only if it is not empty
        if address <> {}:
            node['address'] = address
        #add gnis dict to the node, if not empty
        
        if gnis <> {}:
            node['gnis'] = gnis
        #add pos list to node
        node['pos'] = pos
        #add created dict to node
        node['created'] = created
        #add type to node
        node['type'] = 'node'
        
        return node

    if element.tag == 'way':
        node_refs = []
        
        #iterate for all child tags under top element
        for n in element.iter():
            
            #only 'tag' tags evaluated
            if n.tag == 'tag':
                #assign the k and v values of the tag
                key = n.attrib.values()[0]
                val = n.attrib.values()[1]
                
                #only work on k values with no problem characters, ignore the rest
                if problemchars.search(key) == None:
                    
                    #create the address dictionary, ignore keys with more than 1 colon    
                    if key.startswith('addr:') and key.count(':') < 2:
                        col = key.find(':')
                        address[key[(col + 1):]] = val
                        
                    #create USGS GNIS dictionary, ignore keys iwth more than 1 colon
                    elif key.startswith('gnis:') and key.count(':') < 2:
                        col = key.find(':')
                        gnis[key[(col + 1):]] = val
                        
                    
                    #add the rest of the key pairs to the node
                    elif key.count(':') < 2:
                        node[key] = val
            
            #add the node refs to a list            
            elif n.tag == 'nd':
                node_refs.append(n.attrib['ref'])
        
        #iterate over the attributes of the main element 'way'
        for k,v in element.attrib.iteritems():
            
            #create the 'created' dictionary
            if k in CREATED:
                created[k] = v
                
            #add the rest of the attribute key,value pairs to the node   
            else:
                node[k] = v
        #add the node_refs list to the node
        node['node_refs'] = node_refs
        #add address dict to the node, only if it is not empty
        if address <> {}:
            node['address'] = address
        #add gnis dict to the node, if not empty
        if gnis <> {}:
            node['gnis'] = gnis
        #add created dict to node
        node['created'] = created
        #add type to node
        node['type'] = 'way'
        
        return node
    else:
        return None


def process_map(file_in, pretty = False):
    # You do not need to change this file
    file_out = "{0}.json".format(file_in)
    data = []
    with codecs.open(file_out, "w") as fo:
        for _, element in ET.iterparse(file_in):
            el = shape_element(element)
            if el:
                if 'address' in el.keys():
                    if 'street' in el['address'].keys():
                        el['address']['street'] = update_name(el['address']['street'], mapping)
                        
                data.append(el)
                if pretty:
                    fo.write(json.dumps(el, indent=2)+"\n")
                else:
                    fo.write(json.dumps(el) + "\n")
    return data

def test():
    # NOTE: if you are running this code on your computer, with a larger dataset, 
    # call the process_map procedure with pretty=False. The pretty=True option adds 
    # additional spaces to the output, making it significantly larger.
    data = process_map('/home/gino/Desktop/map', False)
    #pprint.pprint(data)
    
    
if __name__ == "__main__":
    test()
```
